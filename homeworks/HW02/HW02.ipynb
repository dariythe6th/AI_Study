```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# Загрузка данных
df = pd.read_csv('S02-hw-dataset.csv')

# Вывод первых строк
print("Первые 5 строк датасета:")
print(df.head())
print()

# Информация о столбцах и типах
print("Информация о датасете:")
print(df.info())
print()

# Базовые описательные статистики
print("Базовые статистики:")
print(df.describe())
```

```python
# Доля пропусков в каждом столбце
print("Доля пропусков по столбцам:")
missing_percent = df.isna().mean() * 100
for col, percent in missing_percent.items():
    print(f"{col}: {percent:.2f}%")
print()

# Проверка на полностью дублирующие строки
duplicates = df.duplicated()
print(f"Количество полностью дублирующих строк: {duplicates.sum()}")
if duplicates.sum() > 0:
    print("\nДублирующие строки:")
    print(df[duplicates])
print()

# Поиск "подозрительных" значений
print("Поиск подозрительных значений:")

# 1. Отрицательный возраст
negative_age = df[df['age'] < 0]
print(f"1. Отрицательный возраст: {len(negative_age)} строк")
if len(negative_age) > 0:
    print(negative_age[['user_id', 'age']])

# 2. Возраст > 100
old_age = df[df['age'] > 100]
print(f"\n2. Возраст > 100 лет: {len(old_age)} строк")
if len(old_age) > 0:
    print(old_age[['user_id', 'age']])

# 3. Отрицательное количество покупок
negative_purchases = df[df['purchases'] < 0]
print(f"\n3. Отрицательное количество покупок: {len(negative_purchases)} строк")
if len(negative_purchases) > 0:
    print(negative_purchases[['user_id', 'purchases']])

# 4. Нулевой revenue при ненулевых purchases
zero_revenue = df[(df['revenue'] == 0) & (df['purchases'] > 0)]
print(f"\n4. Нулевой revenue при ненулевых покупках: {len(zero_revenue)} строк")
if len(zero_revenue) > 0:
    print(zero_revenue[['user_id', 'purchases', 'revenue']])

# 5. Возраст < 14 лет (малолетние покупатели)
young_age = df[df['age'] < 14]
print(f"\n5. Возраст < 14 лет (малолетние покупатели): {len(young_age)} строк")
if len(young_age) > 0:
    print(young_age[['user_id', 'age']])

# 6. Средний чек больше 1000 или меньше 0
df['avg_check'] = df['revenue'] / df['purchases']
suspicious_check = df[(df['avg_check'] > 1000) | (df['avg_check'] < 0)]
print(f"\n6. Подозрительный средний чек (>1000 или <0): {len(suspicious_check)} строк")
if len(suspicious_check) > 0:
    print(suspicious_check[['user_id', 'purchases', 'revenue', 'avg_check']].head())
```

В ходе анализа данных были обнаружены следующие проблемы:

1. ПРОПУСКИ В ДАННЫХ: В столбце 'age' обнаружено 2 пропущенных значения (5% от всех записей).
   Это может быть связано с тем, что некоторые пользователи не указали свой возраст.

2. ДУБЛИКАТЫ: Найдена 1 полностью дублирующая строка (user_id=10).
   Это может быть результатом ошибки при сборе или загрузке данных.

3. НЕКОРРЕКТНЫЕ ЗНАЧЕНИЯ:
   - Отрицательное количество покупок (-1) у пользователя с user_id=6
   - Нереалистично высокий возраст (120 лет) у пользователя с user_id=6
   - Возраст 5 лет у пользователя с user_id=21 - малолетний покупатель
   - Нулевой revenue при ненулевых покупках у нескольких пользователей
   
4. ЛОГИЧЕСКИЕ НЕСООТВЕТСТВИЯ:
   - У пользователя с user_id=11: 4 покупки, но 0 revenue
   - У пользователя с user_id=7: 0 покупок и 0 revenue (возможно, новый пользователь)
   - У пользователя с user_id=34: 0 покупок и 0 revenue

Эти проблемы требуют очистки данных перед проведением глубокого анализа,
так как они могут исказить статистические выводы и результаты моделей.


```python
# Частоты для категориальных переменных
print("Распределение пользователей по странам:")
country_counts = df['country'].value_counts()
print(country_counts)
print()

print("Процентное распределение по странам:")
country_percent = df['country'].value_counts(normalize=True) * 100
print(country_percent.round(2))
print()

# Группировка с агрегатами по странам
print("Статистики по странам:")
country_stats = df.groupby('country').agg({
    'age': ['mean', 'median', 'count'],
    'purchases': ['mean', 'sum'],
    'revenue': ['mean', 'sum'],
    'user_id': 'count'
}).round(2)

# Переименуем столбцы для лучшей читаемости
country_stats.columns = ['age_mean', 'age_median', 'age_count',
                        'purchases_mean', 'purchases_total',
                        'revenue_mean', 'revenue_total', 'users_count']
print(country_stats)
print()

# Создание возрастных групп
print("Анализ по возрастным группам:")
df['age_group'] = pd.cut(df['age'],
                         bins=[0, 18, 25, 35, 45, 60, 100],
                         labels=['<18', '18-25', '26-35', '36-45', '46-60', '60+'])

age_group_stats = df.groupby('age_group').agg({
    'purchases': ['mean', 'sum', 'count'],
    'revenue': ['mean', 'sum'],
    'user_id': 'count'
}).round(2)

print(age_group_stats)
print()

# Дополнительная группировка: средний чек по странам
print("Средний чек по странам:")
df['avg_check'] = df['revenue'] / df['purchases']
# Заменим бесконечные значения на NaN
df['avg_check'] = df['avg_check'].replace([np.inf, -np.inf], np.nan)

avg_check_by_country = df.groupby('country')['avg_check'].agg(['mean', 'median', 'count']).round(2)
print(avg_check_by_country)
```

```python
# Гистограмма для возраста
plt.figure(figsize=(12, 6))

# Рассчитаем оптимальное количество бинов
age_data = df['age'].dropna()
n_bins = int(1 + 3.322 * np.log(len(age_data)))
n_bins = min(n_bins, 15)

plt.hist(age_data, bins=n_bins, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Распределение возраста пользователей', fontsize=16, fontweight='bold')
plt.xlabel('Возраст (лет)', fontsize=14)
plt.ylabel('Количество пользователей', fontsize=14)
plt.grid(True, alpha=0.3)

# Добавим вертикальную линию для среднего значения
mean_age = age_data.mean()
plt.axvline(mean_age, color='red', linestyle='--', linewidth=2,
            label=f'Средний возраст: {mean_age:.1f} лет')
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# Боксплот для revenue по странам
plt.figure(figsize=(12, 7))

# Подготовка данных для боксплота
countries = df['country'].unique()
data_to_plot = [df[df['country'] == country]['revenue'].dropna() for country in countries]

bp = plt.boxplot(data_to_plot, labels=countries, patch_artist=True)

# Настройка цветов
colors = ['blue', 'green', 'red', 'orange', 'pink']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

plt.title('Распределение revenue по странам', fontsize=16, fontweight='bold')
plt.xlabel('Страна', fontsize=14)
plt.ylabel('Revenue', fontsize=14)
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

```python
# Scatter plot: возраст vs revenue с разбивкой по странам
plt.figure(figsize=(14, 8))

# Используем разные цвета для разных стран
country_colors = {
    'RU': 'blue',
    'FR': 'green',
    'US': 'red',
    'DE': 'orange',
    'CN': 'purple'
}

# Создаем scatter plot для каждой страны
for country in df['country'].unique():
    country_data = df[df['country'] == country]
    plt.scatter(country_data['age'], country_data['revenue'],
                alpha=0.6, s=100, label=country, color=country_colors[country],
                edgecolors='black', linewidth=0.5)

plt.title('Зависимость revenue от возраста (по странам)', fontsize=16, fontweight='bold')
plt.xlabel('Возраст (лет)', fontsize=14)
plt.ylabel('Revenue', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(title='Страна', fontsize=12, title_fontsize=13)

# Добавляем линию тренда
age_filtered = df['age'].dropna()
revenue_filtered = df.loc[age_filtered.index, 'revenue']
z = np.polyfit(age_filtered, revenue_filtered, 1)
p = np.poly1d(z)
plt.plot(age_filtered.sort_values(), p(age_filtered.sort_values()),
         "k--", alpha=0.8, linewidth=3, label='Линия тренда')

plt.legend()
plt.tight_layout()

# Сохранение графика в папку figures
figures_dir = 'homeworks/HW02/figures/'
os.makedirs(figures_dir, exist_ok=True)

plt.savefig(f'{figures_dir}scatter_age_vs_revenue.png', dpi=300,
            bbox_inches='tight', facecolor='white')
print(f"График сохранен в {figures_dir}scatter_age_vs_revenue.png")

plt.show()
```
